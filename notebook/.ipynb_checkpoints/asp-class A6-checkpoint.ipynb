```python
inputFile = '../../sounds/sax-phrase-short.wav'
stdThsld=10
minNoteDur=0.1
winStable = 3
window='hamming'
M=1024
N=2048
H=256
f0et=5.0
t=-100
minf0=310
maxf0=650
```

```python
fs, x = UF.wavread(inputFile)
w  = get_window(window, M)
f0 = HM.f0Detection(x, fs, w, N, H, t, minf0, maxf0, f0et)
```

```python
f0[f0==0] = eps
f0c = 1200 * np.log2(f0 / 55.0)
```

```python
means2 = np.zeros(f0c.size-winStable+1)
for i in xrange(winStable):
    means2 += f0c[i:f0c.size-winStable+i+1]
means2 /= winStable

std2 = np.zeros(f0c.size-winStable+1)
for i in xrange(winStable):
    std2 += np.power(f0c[i:f0c.size-winStable+i+1]-means2, 2)
std2 = np.sqrt(std2/winStable)

means = np.zeros(f0c.size)
means[winStable-1:] = np.array([f0c[i:i+winStable].mean() for i in xrange(f0c.size-winStable+1)])
means[:winStable-1] = nan

std = np.zeros(f0c.size)
std[winStable-1:] = np.array([f0c[i:i+winStable].std() for i in xrange(f0c.size-winStable+1)])
std[:winStable-1] = nan

#np.where(std < stdThsld)
ends = np.where((std[:-1] < stdThsld) & (std[1:] >= stdThsld))[0]
starts = np.where((std[:-1] >= stdThsld) & (std[1:] < stdThsld))[0]+1
segments = np.array([starts, ends])
segments
```

```python
min_segment_size = minNoteDur / (x.size / fs / f0.size)
```

```python
print (segments[1,:] - segments[0,:])
print (segments[1,:] - segments[0,:]) > min_segment_size
full_segments = segments[:,(segments[1,:] - segments[0,:]) > min_segment_size]
np.transpose(full_segments)
```
